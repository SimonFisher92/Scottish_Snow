```python
import cv2
import numpy as np
import os
import random

def process_image(img_path, mask_dir):
    global img, mask

    def draw(event, x, y, flags, param):
        global drawing, points

        if event == cv2.EVENT_LBUTTONDOWN:
            drawing = True
            points = [(x, y)]

        elif event == cv2.EVENT_MOUSEMOVE:
            if drawing:
                points.append((x, y))
                cv2.line(mask, points[-2], points[-1], (0, 0, 255), thickness=5)

        elif event == cv2.EVENT_LBUTTONUP:
            drawing = False
            cv2.fillPoly(mask, [np.array(points, dtype=np.int32)], (0, 0, 255))

    img = cv2.imread(img_path)
    mask = np.zeros_like(img, dtype=np.uint8)

    cv2.namedWindow('image')
    cv2.setMouseCallback('image', draw)

    while True:
        combined = cv2.addWeighted(img, 0.85, mask, 0.85, 0)
        cv2.imshow('image', combined)

        k = cv2.waitKey(1) & 0xFF
        if k == 27:  # ESC key
            break

    base, _ = os.path.splitext(os.path.basename(img_path))
    mask_output_path = os.path.join(mask_dir, base + 'gtmask.png')

    cv2.imwrite(mask_output_path, mask)
    cv2.destroyAllWindows()

# Directory containing the images
directory = img_path
mask_directory = os.path.join(directory, 'ground_truth_masks')

# Create the subdirectory for ground truth masks if it doesn't exist
if not os.path.exists(mask_directory):
    os.makedirs(mask_directory)

all_files = [os.path.join(directory, f) for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

selected_files = random.sample(all_files, len(all_files) // 2)

for file in selected_files:
    process_image(file, mask_directory)
```
